```python
import matplotlib.pyplot as plt
```

```python
categories = ["Food", "Transportation", "Shopping", "Health", "Personal", "Other"]
```

```python
import pandas as pd
expenses = {cat: 0.0 for cat in categories}

print("Enter your expenses (type 'quit' to finish):")
while True:
    category = input(f"Choose a category {categories[:-1]} (or type 'quit' to finish): ")

    if category.lower() == "quit":
        break

    amount = input("Enter amount: ")
    try:
        amount = float(amount)
    except ValueError:
        print(" Please enter a valid number for amount.")
        continue

    # If category is valid → add to that category
    if category in categories[:-1]:
        expenses[category] += amount
    else:
        # If category doesn't match → go into "Other"
        expenses["Other"] += amount
```

```python
print("\nTotal expenses by category:")
df = pd.DataFrame(list(expenses.items()), columns=["Category", "Total ($)"])
print(df.to_string(index=False))

```

```python
plt.bar(expenses.keys(), expenses.values(), color="green")
plt.xlabel("Category")
plt.ylabel("Total Spent ($)")
plt.title("Expenses by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
import csv
with open("expenses.csv", mode="w", newline="") as file:
    writer=csv.writer(file)
    writer.writerow(["Category", "Amount"])
    for category, amount in expenses.items():
        writer.writerow([category, amount])
print("Expenses saved to expenses.csv")
```

```python
loaded_expenses={}
with open("expenses.csv",mode="r") as file:
    reader=csv.DictReader(file)
    for row in reader:
        loaded_expenses[row["Category"]]= float(row["Amount"])
print("Loaded expenses:", loaded_expenses)
```
