```python
import matplotlib.pyplot as plt
```

```python
categories = ["Food", "Transportation", "Shopping", "Health", "Personal", "Other"]
```

```python
expenses= {category: 0 for category in categories}
while True:
    category=input(f"Choose category {categories} (or type 'quit' to finish):")
    if category.lower()== 'quit':
        break
    try:
        amount = float(input("Enter amount: "))
    except ValueError:
        print("Invalid amount. PLease enter a number.")
        continue
    if category in expenses:
        expenses[category]+= amount
    else:
        print("Invalid category, added to 'Other'")
        expenses ["Other"] += amount
```

```python
print("Total\\expense by category:")
for category, amount in expenses.items():
    print(f"{category}: ${amount:.2f}")
```

```python
plt.pie(expenses.values(), labels=expenses.keys(), autopct='%1.1f%%')
plt.title("Expenses by Category")
plt.show()
```

```python
import csv
with open("expenses.csv", mode="w", newline="") as file:
    writer=csv.writer(file)
    writer.writerow(["Category", "Amount"])
    for category, amount in expenses.items():
        writer.writerow([category, amount])
print("Expenses saved to expenses.csv")
```

```python
loaded_expenses={}
with open("expenses.csv",mode="r") as file:
    reader=csv.DictReader(file)
    for row in reader:
        loaded_expenses[row["Category"]]= float(row["Amount"])
print("Loaded expenses:", loaded_expenses)
```
